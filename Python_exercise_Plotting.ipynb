Task-1 Find mean, median and mode

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/grlearningnew2.csv')
print(data)

mean1 = data['Petal.Width'].mean()
mean2 = data['Petal.Length'].mean()
mean3 = data['Sepal.Length'].mean()
print('Mean of petal width: ',str(mean1))
print('Mean of petal length: ' + str(mean2))
print('Mean of setal length: ' + str(mean3))

median1 = data['Petal.Width'].median()
median2 = data['Petal.Length'].median()
median3 = data['Sepal.Length'].median()
print('Median of petal width: ' + str(median1))
print('Median of petal length: ' + str(median2))
print('Median of sepal length: ' + str(median3))


mode1 = data['Petal.Width'].mode()
mode2 = data['Petal.Length'].mode()
mode3 = data['Sepal.Length'].mode()
print('Mode of petal width: ', mode1)
print('Mode of petal length: ', mode2)
print('Mode of sepal length: ',mode3)
```

Task-3 Plotting outcome of a game

```python
import random
import matplotlib.pyplot as plt

gamelist = ['rock', 'paper', 'scissors']
comp_wins = 0
user_wins = 0
total_games = 0

num_games = int(input("Enter the number of games: "))
success_rate_user = []
success_rate_comp = []

for _ in range(num_games):
    comp = random.choice(gamelist)
    user = input("Enter choice of user: ")

    total_games += 1

    print("The choice of computer:", comp)

    if (comp == "rock" and user == "paper") or (comp == "paper" and user == "scissors") or (comp == "scissors" and user == "rock"):
        user_wins += 1
    elif (user == "rock" and comp == "paper") or (user == "paper" and comp == "scissors") or (user == "scissors" and comp == "rock"):
        comp_wins += 1

    success_rate_user.append(user_wins / total_games)
    success_rate_comp.append(comp_wins / total_games)

    print("User wins:", user_wins)
    print("Computer wins:", comp_wins)
    print("Ties:", total_games - (user_wins + comp_wins))
    print()

plt.plot(success_rate_user, label='User Success Rate', color='blue')
plt.plot(success_rate_comp, label='Computer Success Rate', color='red')
plt.xlabel('Games Played')
plt.ylabel('Success Rate')
plt.title('User vs Computer Success Rate')
plt.legend()
plt.show()
```
