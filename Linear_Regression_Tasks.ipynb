```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt


#x=np.array()
for i in range(10):
  x=np.array(random.randint(1,10)).reshape((-1,1))
  print(x)
  y=np.array(random.randint(11,20)).reshape((-1,1))
  print(y)
  print()
  plt.scatter(x,y)
  #plt.plot(x,pred,color='Red', marker='*')
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()



from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Red', marker='*')


print(type(x))

print("MSE:",mse)
print("intercept", SLR.intercept_)
print("slope: ",SLR.coef_)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from array import *
a=[]
b=[]
n=int(input("Enter the number of elements"))
for i in range(n):
    c=np.random.randint(15)
    d=np.random.randint(15)
    a.append(c)
    b.append(d)
print(a)
print(b)
x=np.array(a).reshape((-1,1)) #To convert a list to array
y=np.array(b)
#print(x)
#print(y)
plt.scatter(x,y)
plt.xlabel('IV')
plt.ylabel('DV')
plt.show()
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Red',marker='*')
plt.show()
print("MSE:",mse)
print("INTERCEPT:",SLR.intercept_)
print("slope:",SLR.coef_)
```
